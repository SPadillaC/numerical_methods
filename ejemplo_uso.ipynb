## Ejemplo de uso de la librer√≠a Numerical Methods

```python
import numpy as np
import matplotlib.pyplot as plt
from numerical_methods import euler_method, central_difference, integral_riemann, integral_trapezoidal

# Definir la ecuaci√≥n diferencial
def f(t, y):
    if y > 1e5:  # Limitar el valor de y para evitar overflow
        return np.inf
    return y**2 * np.exp(t)

# Condiciones iniciales
y0 = 0.5
t0 = 0
tf = 1.4  # Ajuste de l√≠mite superior
h = 0.01  # Reducir tama√±o del paso

# Soluci√≥n anal√≠tica
def analytical_solution(t):
    C = 1 / y0 + 2  # Ajuste basado en la condici√≥n inicial
    return 1 / (C - np.exp(t))

# Resolver la ecuaci√≥n diferencial usando el m√©todo de Euler
t_values, y_values = euler_method(f, y0, t0, tf, h)

# Calcular la soluci√≥n anal√≠tica
t_analytical = np.linspace(t0, tf, 100)
y_analytical = analytical_solution(t_analytical)

# Graficar las soluciones
plt.figure(figsize=(10, 6))
plt.plot(t_values, y_values, 'o-', label='M√©todo de Euler')
plt.plot(t_analytical, y_analytical, '-', label='Soluci√≥n Anal√≠tica')
plt.xlabel('t')
plt.ylabel('y')
plt.xlim([0, 1.39])
plt.ylim([0, 10])
plt.title('M√©todo de Euler vs Soluci√≥n Anal√≠tica')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Definir la funci√≥n
def f(x):
    return np.cos(x)

# Punto en el que se calcular√° la derivada
x = np.pi / 4

# Calcular la derivada central
derivada_central = central_difference(f, x)
print(f"La derivada central de f(x) en x={x} es: {derivada_central}")

# Definir los l√≠mites de integraci√≥n
a = 0
b = np.pi / 2
n = 1000  # n√∫mero de subintervalos

# Calcular la integral usando los m√©todos de Riemann y trapezoidal
I_riemann = integral_riemann(f, a, b, n)
I_trapezoidal = integral_trapezoidal(f, a, b, n)

print(f"La integral calculada num√©ricamente con el m√©todo de Riemann es: {I_riemann}")
print(f"La integral calculada num√©ricamente con el m√©todo trapezoidal es: {I_trapezoidal}")
```
